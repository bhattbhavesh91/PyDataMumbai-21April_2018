```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_RGBimage(image):
    '''
    OpenCV represents RGB images as multi-dimensional NumPy array but in reverse order.
    This means that images are actually represented in BGR order rather than RGB.
    '''
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis("on")
    return plt.show()
```

```python
def plot_histogram(image):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    plt.figure()
    plt.title("Grayscale Histogram")
    plt.xlabel("Bins")
    plt.ylabel("# of Pixels")
    plt.plot(hist)
    plt.xlim([-10, 260])
    return plt.show()
```

## Loading an image

```python
image = cv2.imread("pydata_mumbai.jpg")
```

## Displaying the inputted image

```python
show_RGBimage(image)
```

### Creating a black image

```python
blank_image = np.zeros((400,400,3))
type(blank_image[0,0,0])
```

```python
black_image = np.zeros((400,400,3), np.uint8)
show_RGBimage(black_image)
```

### Creating a white image

```python
white_image = 255*np.ones((400,400,3), np.uint8)
show_RGBimage(white_image)
```

```python
image_ref1 = image
```

```python
image_ref1[0:100, 0:100] = (0, 0, 0)
show_RGBimage(image_ref1)
```

## Drawing shapes using OpenCV

```python
canvas = 255*np.ones((300, 300, 3), dtype = "uint8")
red    = (0, 0, 255)  #OpenCV representation of Red is based on BGR and not RGB representation
# Draw a rectangle starting from point (50,50) till (150,250) the points are represented as (x,y)
cv2.rectangle(canvas, (50, 50), (150,250), red, 5);
```

```python
show_RGBimage(canvas)
```

### Masking

```python
mask = np.zeros(image.shape[:2], dtype = "uint8")
cv2.rectangle(mask, (50, 150), (100, 375), 255, -1);
# By specifying -1 as the thickness, our rectangle is drawn as a solid white line.
masked = cv2.bitwise_and(image, image, mask = mask)
show_RGBimage(masked)
```

## Histograms
- Displays the frequency distribution of the intensity values of an image

```python
image_white_gray = cv2.cvtColor(white_image, cv2.COLOR_BGR2GRAY)
plot_histogram(image_white_gray)
```

```python
image_black_gray = cv2.cvtColor(black_image, cv2.COLOR_BGR2GRAY)
plot_histogram(image_black_gray)
```
